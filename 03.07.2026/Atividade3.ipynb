# Questão 1
Crie quatro variáveis para armazenar as seguintes informações sobre um produto: 

- Nome: "NVIDIA GeForce RTX 5090" (tipo texto). 
- Quantidade de RAM: 32 (tipo inteiro). 
- Preço: 10.871,36 (tipo decimal). 
- Em estoque: True (tipo booleano). 

Após criar as variáveis, utilize a função `print()` e `type()` para exibir o valor e o tipo de cada uma delas no console.

```python
name = "NVIDIA GeForce RTX 5090"
version = 32
price = 10.871
Stock = True

list = [name, version, price, Stock]

for _ in list:
    print(_)
    print(type(_))
    
```

# Questão 2
Imagine que você está preparando o seu ambiente de programação e organizando as bibliotecas que usará para manipulação e visualização de dados.

3. Adicione o pacote "shiny" ao final da lista.

# 🏆 Desafio Final - Frota Logística

Você foi contratado para criar o sistema interno de um aplicativo de entregas. Sua missão é organizar as informações dos veículos disponíveis utilizando as estruturas de dados adequadas que aprendeu hoje:

1. Última Rota (Tupla): Crie uma tupla imutável para guardar as coordenadas geográficas (Latitude e Longitude) do último ponto de entrega do veículo (ex: (-7.115, -34.863)).

2. Ficha do Veículo (Dicionário): Cada veículo deve ser representado por um dicionário contendo as seguintes chaves:

- modelo (str)

- ano (int)

- capacidade_kg (float)

- em_operacao (bool)

- ultima_coordenada (a tupla criada no passo 1)

3- Frota Ativa (Lista): Agrupe os dicionários de pelo menos dois veículos dentro de uma única lista chamada frota_aplicativo.

```python
coordenadas = (-7.115, -34.863)
ficha_veiculo1 = {
    "modelo":"Meriva",
    "ano":2010,
    "capacidade":10.5,
    "em_operacao":True,
    "ultima_coordenada":coordenadas,
     
}

ficha_veiculo2 = {
    "modelo":"Tucson",
    "ano":2012,
    "capacidade":9.5,
    "em_operacao":False,
    "ultima_coordenada":coordenadas,
     
}

frota_aplicativo = []

frota_aplicativo.append(ficha_veiculo1)
frota_aplicativo.append(ficha_veiculo2)

print(frota_aplicativo)

```
